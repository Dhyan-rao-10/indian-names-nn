```python
import pandas as pd

df = pd.read_csv('../data/Indian_Names.csv')
print(df.columns.tolist())
print(df.head())
```

```python
names = df['Name'].dropna().str.lower().str.strip()
names = names[names.str.match(r'^[a-z]+$')]  # keep only plain alphabet names
print(f"Total names: {len(names)}")
print(names.head(20))
```

```python
names.to_csv('../data/names.txt', index=False, header=False)
print("Done! names.txt saved.")
```

```python
print(f"Total names: {len(names)}")
print(f"Shortest name: {names.str.len().min()} chars")
print(f"Longest name: {names.str.len().max()} chars")
print(f"Average length: {names.str.len().mean():.1f} chars")
print("\nSample of 10 random names:")
print(names.sample(10).tolist())
```
